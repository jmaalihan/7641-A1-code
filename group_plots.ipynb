```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
clfs = ['MLP', 'DT', 'GBDT', 'KNN', 'SVM']
scores = {
    'Iris': (97.33, 96.67, 96.67, 96.31, 97.33),
    'Breast Cancer': (97.54, 94.90, 97.36, 96.30, 97.71)
}

x = np.arange(len(clfs))  # the label locations
width = 0.4  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in scores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy')
ax.set_title('Test Scores')
ax.set_xticks(x + width, clfs)
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(90, 100)

plt.show()
```

```python
clfs = ['MLP', 'DT', 'GBDT', 'KNN', 'SVM']
scores = {
    'Accuracy': (97.54, 94.90, 97.36, 96.30, 97.71),
    'F1': (98.05, 95.93, 97.93, 97.13, 98.20)
}

x = np.arange(len(clfs))  # the label locations
width = 0.4  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')
colors = ['palegreen', 'pink']

cindex = 0
for attribute, measurement in scores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color = colors[cindex % 2])
    cindex +=1
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy vs F1 (Breast Cancer)')
ax.set_xticks(x + width, clfs)
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(90, 100)


plt.show()
```
